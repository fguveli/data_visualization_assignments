1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)

title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }
```

```python
countries = ['Spain', 'France', 'Germany', 'Italy']
linedf = df[(df['location'].isin(countries)) & (df['date'] > '2020-04-01')]
plt.figure(figsize=(10,4))
sns.lineplot(x= linedf.date, y=linedf.total_deaths, data=linedf, hue=linedf.location)
plt.xticks(rotation=45, size=9)
plt.show()
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
bardf = df[df['date'] == '2020-08-13']
plt.figure(figsize=(10,6))
plt.title('Total Cases per Continent', fontdict=title_style)
sns.barplot(x=df.continent, y=bardf.total_deaths, data=bardf)
plt.xticks(rotation=90)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
histdf = df[df['location'] == 'South Africa']
plt.figure(figsize=(16,12))
plt.title('Daily deaths for South Africa')

plt.subplot(2,2,1)
sns.distplot(histdf.new_deaths, color = '#22B573')
plt.title('New Deaths\n10 bins, With Kde')

plt.subplot(2,2,2)
sns.distplot(histdf.new_deaths, color = 'red', kde=False)
plt.title('New Deaths\n10 bins, Without Kde')


plt.subplot(2,2,3)
sns.distplot(histdf.new_deaths,bins=50, color = '#22B573')
plt.title('New Deaths\n50 bins, With Kde')

plt.subplot(2,2,4)
sns.distplot(histdf.new_deaths, color = 'red', bins=50, kde=False)
plt.title('New Deaths\n50 bins, Without Kde')


plt.show()

```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
countries2 = ['United States', 'Brazil']
boxdf = df[df['location'].isin(countries2)]

plt.figure(figsize=(16, 6))

plt.subplot(131)
sns.boxplot(x= boxdf.location, y=boxdf.new_deaths, data=boxdf, palette = ['#22B573', '#FF914D'])
plt.title('Box Plot', fontdict=title_style)

plt.subplot(132)
sns.violinplot(x=boxdf.location, y=boxdf.new_deaths, palette = ['#22B573', '#FF914D'])
plt.title('Violin Plot', fontdict=title_style)

plt.subplot(133)
sns.swarmplot(x= boxdf.location, y=boxdf.new_deaths, data=boxdf, palette = ['#22B573', '#FF914D'])
plt.title('Swarm Plot', fontdict=title_style)

plt.show()
```
