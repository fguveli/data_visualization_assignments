In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
from datetime import date, timedelta
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)

countryList = ['Spain', 'France', 'Germany', 'Italy']
mydf = df[df['location'].isin(countryList)]
#lastDay = pd.date_range(end=date.today(), periods = 100)
lastDay = pd.date_range(end='2020-08-14', periods = 100)
lastDay = (lastDay[0])
mydf = mydf[mydf['date'] > lastDay]
```

```python
fig = plt.figure(figsize=(16,8)) 

plt.subplot(1,2,1)
#loop for each country., with mode operation changing linewidth
for cnt in countryList:
    tmpdf = mydf[mydf['location'] == cnt]
    plt.plot(tmpdf['date'],tmpdf['total_cases'], label=cnt, linewidth = (ord(cnt[1]) % 5))
    
#x axis data labels
plt.xticks(rotation = 45, fontsize=11)    

plt.title('Total Cases per Location',
         fontsize= 20,
         color = 'red')
plt.xlabel('Date',
          fontstyle = 'normal', 
          fontsize = 15,
          fontweight = 'normal', 
          color = "darkblue")
plt.ylabel('Number of cases',
           fontstyle = 'normal', 
          fontsize = 15,
          fontweight = 'normal', 
          color = "darkblue")
plt.legend(title = 'Countries', loc=4)

plt.subplot(1,2,2)
#loop for each country.
for cnt in countryList:
    tmpdf = mydf[mydf['location'] == cnt]
    plt.plot(tmpdf['date'],tmpdf['total_deaths'], label=cnt, linewidth = (ord(cnt[1]) % 5))

plt.xticks(rotation = 45, fontsize=11)
#renaming chart title
plt.title('Total Deaths per Location',
         fontsize= 20,
         color = 'red')
#renaming x axis title
plt.xlabel('Date',
           fontstyle = 'normal', 
          fontsize = 15,
          fontweight = 'normal', 
          color = "darkblue",)
plt.ylabel('Number of deaths',
           fontstyle = 'normal', 
          fontsize = 15,
          fontweight = 'normal', 
          color = "darkblue")
plt.legend(title = 'Countries', loc=4)

fig.suptitle('Covid-19 in Europe', fontsize = 24, color='green')
plt.show()
```
